```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import sklearn as sk
```

```python
df = pd.read_csv("D:\Cognifyz Internship\Dataset .csv")
df
```

```python
lab=LabelEncoder()
```

```python
df['Rating text']=lab.fit_transform(df['Rating text'])
df
```

```python
max_votes_restaurant = df.loc[df['Votes'].idxmax()]
min_votes_restaurant = df.loc[df['Votes'].idxmin()]
```

```python
print(f"Highest votes restaurant: {max_votes_restaurant['Restaurant Name']} ({max_votes_restaurant['Votes']} votes)")
print(f"Lowest votes restaurant: {min_votes_restaurant['Restaurant Name']} ({min_votes_restaurant['Votes']} votes)")
```

```python
correlation = df['Votes'].corr(df['Rating text'])
correlation
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Votes', y='Rating text', data=df)
plt.xlabel("Number of Votes")
plt.ylabel("Rating")
plt.title("Votes vs. Rating")
plt.show()
```
